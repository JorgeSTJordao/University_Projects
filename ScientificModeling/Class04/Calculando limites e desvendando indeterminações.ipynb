### Limite laterais

```python
def limitesLaterais(f, x, x_limite):

    def valoresDeAproximacao(x_limite, distancia=[0.1, 0.01, 0.001, 0.0001, 0.00001]):

        pontos_esquerda = [x_limite - d for d in distancia]
        pontos_direita = [x_limite + d for d in distancia]

        return pontos_esquerda, pontos_direita

    y_limite = limit(f, x, x_limite)
    
    print(f"Valor limite {y_limite}")
    
    pontos_esquerda, pontos_direita = valoresDeAproximacao(x_limite)

    pontos_esquerda_y = [f.subs(x, x_valor) for x_valor in pontos_esquerda]
    pontos_direita_y = [f.subs(x, x_valor) for x_valor in pontos_direita]

    print(f"Aproximações pela esquerda: {pontos_esquerda_y}")
    print(f"Aproximações pela direita: {pontos_direita_y}")
```

```python
def fun_val(f, x_valor):
    return f.subs(x, x_valor).evalf()

x = symbols('x')
x_valores = np.linspace(1, 5, 5)
f = x**2

y = []
for x_value in x_valores:
    y.append(fun_val(f, x_value)) 
    
print(y)
```

### Plotando o Gráfico

```python
def grafico(f, x, x_limite):
    
    #Cálculo de x, desconsiderando qualquer tipo de indeterminação
    def fun_val(f, x_valor):
        return f.subs(x, x_valor).evalf()

    #Limite
    x_inicio = x_limite - 1
    x_final = x_limite + 1

    #Limite da função
    y_limite = limit(f, x, x_limite)
    
    #Valores próximos do limite
    x_vals = np.linspace(x_inicio, x_final, 500)
    x_vals = x_vals[x_vals != x_limite]

    #Cálculo de y quando se aproxima do limite
    y_vals = []
    for x_valor in x_vals:
        y_vals.append(fun_val(f, x_valor))

    #Gráfico de função
    plt.plot(x_vals, y_vals, label='{}'.format(f)) 
    
    #Local em que ocorre a descontinuidade
    plt.scatter([x_limite], [y_limite], s=30, facecolors='white', edgecolors='black')
    
    #Linha tracejada na vertical para localizar o limite
    plt.plot([x_limite, x_limite], [0, y_limite], 'orange', linestyle='--')
    
    #Linha tracejada na horizontal para localizar o limite
    plt.plot([x_inicio, x_limite], [y_limite, y_limite], 'orange', linestyle='--')    
    
    #Quadriculado
    plt.grid()
    
    #Legenda para representar a função
    plt.legend(loc='best')
```
