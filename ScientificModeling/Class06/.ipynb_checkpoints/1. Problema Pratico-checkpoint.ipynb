### Concavidade

```python
def f(x):
    return x**4 - 20*x**3 + 100*x**2 + 100

def g(x):
    return 12*x**2 - 120*x + 200

def z(x):
    return 24*x - 120

maior = None
menor = None

for x in range(-4, 11):

    if g(x) > 0:
        val = g(x)
        print(f"Mínimo local: {val}. Ponto {x}")
    else:
        val = g(x)
        print(f"Máximo local: {val}. Ponto {x}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 20, 1000)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, z(x))
plt.ylim(-250, 1000)
plt.grid()
```
