## Polinômio de Langrage com Scipy

### Bibliotecas principais

```python
import numpy as np #Arrays a serem trabalhadas
import matplotlib.pyplot as plt #Plotagem de Gráfico
from scipy.interpolate import lagrange #Função de langrage
from numpy.polynomial.polynomial import Polynomial #Detectando a função polinomial
```

```python
#Arrays do tipo numpy
x = np.array([300, 304, 305, 307])
y = np.array([2.4471, 2.4829, 2.4843, 2.4871])
```

```python
#Calculando a função polinomal (Grau 3)
polinomio = lagrange(x, y)
print(polinomio)

#Out: 0.0002157 x**3 - 0.1976 x**2 + 60.33 x - 6138
```

```python
#Imprimindo os valores dos polinômios em ordem decrescente
print(Polynomial(polinomio.coef[::-1]).coef)
```

```python
#Imprindo o gráfico
import matplotlib.pyplot as plt
plt.scatter(x, y, label='Pontos')

plt.legend()
plt.show()
```
