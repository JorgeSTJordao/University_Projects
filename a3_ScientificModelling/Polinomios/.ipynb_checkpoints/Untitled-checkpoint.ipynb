```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Latex
```

```python
x = sp.symbols('x')

#Função no sympy
f_sympy = x**3 - x**2 + 1

#display(Latex('$f(x) = x^3 - x^2 + 1$'))

#Função derivada no sympy
f_sympy_diff = sp.diff(f_sympy)

#display(Latex('$f(x)^` = 3x^2 - 2x$'))

#Função derivada duas vezes no sympy
f_sympy_diff2 = sp.diff(f_sympy_diff)

#display(Latex('$f(x)^` = 6x - 2$'))

# Convertendo expressão para função Numpy
p = sp.lambdify(x, f_sympy)
p_diff = sp.lambdify(x, f_sympy_diff)
p_diff2 = sp.lambdify(x, f_sympy_diff2)

# Criando um array de pontos
x_p = np.linspace(-200, 200, 500)

# Calculando valores da função
y = p(x_p)

# Calculando valores da função derivada
y_diff = p_diff(x_p)

# Calculando valores da função derivada duas vezes
y_diff2 = p_diff2(x_p)

plt.grid()

plt.plot(x_p, y)
plt.plot(x_p, y_diff)
plt.plot(x_p, y_diff2)
```
