### Trançando o gráfico

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
#1ª Função
ax1 = plt.subplot(2, 2, 1)
x = np.linspace(0, 100, 1000)
y = x**2

ax1.plot(x, y, color='red')

#2ª Função
ax2 = plt.subplot(2, 2, 2)
#Intervalo definido
x = np.linspace(-2, 2, 1001)
y = 2*x

ax2.plot(x, y, color='orange')

#3ª Função
ax3 = plt.subplot(2, 2, 3)
x = np.linspace(0, 2*np.pi)
y = np.cos(x)

ax3.plot(x, y, color='green')

#4ª Função
ax4 = plt.subplot(2, 2, 4)
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

ax4.plot(x, y, color='blue')
```

### Triângulo retângulo

```python
plt.xlim(-5, 5)
plt.ylim(-7, 7)


plt.grid()

ct_ad = {'x': [-3, 3], 'y': [-2, -2]}
ct_op = {'x': [3, 3], 'y': [-2, 5]}

color=''

plt.plot(ct_ad['x'], ct_ad['y'], label='cateto adjacente')
plt.plot(ct_op['x'], ct_op['y'], label='cateto oposto')

distance_x = abs(ct_ad['x'][0]) + (ct_ad['x'][1])
distance_y = abs(ct_op['y'][0]) + (ct_op['y'][1])

plt.plot(ct_ad['x'], ct_op['y'], label='hipotenusa')

plt.legend(loc='best')
```

### Gráfico definido

```python
plt.ylim(0, 5)

#Intervalo 1
x = np.linspace(-100, 1, 1000)
y = (x - x) + 1

plt.plot(x, y, label='f(x) = (x - x) + 1')

x = np.linspace(1, 2, 1000)[1:]
y = 2*x

plt.plot(x, y, label='f(x) = 2*x')

x = np.linspace(2, 100, 1000)[1:]
y = (x - x) + 4

plt.plot(x, y, label='f(x) = (x - x) + 4')

plt.grid()
plt.legend(loc='best')
```

### Polígonos regulares

```python
from matplotlib.patches import RegularPolygon

plt.xlim([-2, 2])
plt.ylim([-2, 2])

poligono = RegularPolygon((0, 0), 20, radius=1, edgecolor='orange')

#Adicionando a figura ao plot
plt.gca().add_patch(poligono)

plt.grid()
```
