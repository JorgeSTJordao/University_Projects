### Continuidade com erro do Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Funo contnua 1: f(x) = x^2
def f1(x):
    return x ** 2

# Funo contnua 2: f(x) = sin(x)
def f2(x):
    return np.sin(x)

# Funo descontnua 1: Funo de Heaviside
def f3(x):
    return np.heaviside(x, 0.5)

# Funo descontnua 2: Parte inteira (floor function)
def f4(x):
    return np.floor(x)

# Definindo os valores de x
x_cont = np.linspace(-10, 10, 400)
x_disc = np.linspace(-10, 10, 400)

# Criando figuras e eixos para os grficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plotando as funes contnuas
axs[0, 0].plot(x_cont, f1(x_cont), 'r')
axs[0, 0].set_title('Funo Contnua $f(x) = x^2$')

axs[0, 1].plot(x_cont, f2(x_cont), 'b')
axs[0, 1].set_title('Funo Contnua $f(x) = \sin(x)$')

# Plotando as funes descontnuas
axs[1, 0].plot(x_disc, f3(x_disc), 'g')
axs[1, 0].set_title('Funo Descontnua: Heaviside')

axs[1, 1].plot(x_disc, f4(x_disc), 'm')
axs[1, 1].set_title('Funo Descontnua: Parte inteira $\\lfloor x \\rfloor$')

# Ajustando o layout
plt.tight_layout()

# Mostrando os grficos
plt.show()
```

### Continuidade ajeitada

```python
import matplotlib.pyplot as plt
import numpy as np

# Funo contnua 1: f(x) = x^2
def f1(x):
    return x ** 2

# Funo contnua 2: f(x) = sin(x)
def f2(x):
    return np.sin(x)

# Funo descontnua 1: Funo de Heaviside
def f3(x):
    return np.heaviside(x, 0.5)

# Funo descontnua 2: Parte inteira (floor function)
def f4(x):
    return np.floor(x)

# Definindo os valores de x
x_cont = np.linspace(-10, 10, 400)
x_disc = np.linspace(-10, 10, 400)

# Criando figuras e eixos para os grficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plotando as funes contnuas
axs[0, 0].plot(x_cont, f1(x_cont), 'r')
axs[0, 0].set_title('Funo Contnua $f(x) = x^2$')

axs[0, 1].plot(x_cont, f2(x_cont), 'b')
axs[0, 1].set_title('Funo Contnua $f(x) = \sin(x)$')

# Plotando as funes descontnuas
# Para evitar a conexo entre os pontos descontnuos, utilizaremos o parmetro 'drawstyle'
axs[1, 0].plot(x_disc, f3(x_disc), 'g', linestyle='none', marker='.')
axs[1, 0].set_title('Funo Descontnua: Heaviside')

axs[1, 1].plot(x_disc, f4(x_disc), 'm', linestyle='none', marker='.')
axs[1, 1].set_title('Funo Descontnua: Parte inteira $\\lfloor x \\rfloor$')

# Ajustando o layout
plt.tight_layout()

# Mostrando os grficos
plt.show()
fig.savefig('graficos.png')
```
