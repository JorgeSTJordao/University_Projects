```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(v):
    return 0.00015*v**3 - 0.032*v**2+0.5*v+100

v_min = -2
v_max = 150

v = np.linspace(v_min, v_max, 1000)
y = f(v)
```

```python
plt.plot(v, y, label='f(x) = 0.00015*v**3 - 0.032*v**2+1.8*v+1.7')

pontos = {"Índices": [], "Valores": []}

#Verificação de pontos
for i in range(1, len(y) - 1, 1):
    
    #Ponto máximo
    if y[i - 1] < y[i] and y[i] > y[i + 1]:
        pontos['Índices'].append(i)
        pontos['Valores'].append(y[i])
        
    #Ponto mínimo
    elif y[i - 1] > y[i] and y[i] < y[i+1]:
        pontos['Índices'].append(i)
        pontos['Valores'].append(y[i])
        

if len(pontos['Valores']) >=2:
    i_max = pontos['Valores'].index(max(pontos['Valores']))
    i_min = pontos['Valores'].index(min(pontos['Valores']))

    v_max = v[pontos['Índices'][i_max]]
    v_min = v[pontos['Índices'][i_min]]

    plt.scatter(v_max, max(pontos['Valores']), c='red', label='f_max({}, {})'.format(round(v_max, 2), round(max(pontos['Valores']), 2)))
    plt.scatter(v_min , min(pontos['Valores']), c='orange', label='f_min({}, {})'.format(round(v_min, 2), round(min(pontos['Valores']), 2)))

    plt.legend(loc='best')

    plt.grid()
else:
    print("Sem máximos e mínimos")
```
