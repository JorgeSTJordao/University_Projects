```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(v):
    return 0.00015*v**3 - 0.032*v**2+1.8*v+1.7

v_min = -2
v_max = 160

v = np.linspace(v_min, v_max, 1000)
y = f(v)
```

```python
plt.plot(v, y, label='f(x) = 0.00015*v**3 - 0.032*v**2+1.8*v+1.7')


for i in range(1, len(y) - 1, 1):
    if y[i - 1] < y[i] and y[i] > y[i + 1]:
        max_y = y[i]
        v_imax = i
    
    if y[i - 1] > y[i] and y[i] < y[i+1]:
        min_y = y[i]
        v_imin = i

plt.scatter(v[v_imax], max_y, c='red', label='f_max({}, {})'.format(round(v[v_imax], 2), round(max_y, 2)))
plt.scatter(v[v_imin], min_y, c='orange', label='f_min({}, {})'.format(round(v[v_imin], 2), round(min_y, 2)))

plt.legend(loc='best')

plt.grid()
```
