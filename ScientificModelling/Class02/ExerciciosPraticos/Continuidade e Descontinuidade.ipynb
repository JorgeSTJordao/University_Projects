### Continuidade com erro do Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Função contínua 1: f(x) = x^2
def f1(x):
    return x ** 2

# Função contínua 2: f(x) = sin(x)
def f2(x):
    return np.sin(x)

# Função descontínua 1: Função de Heaviside
def f3(x):
    return np.heaviside(x, 0.5)

# Função descontínua 2: Parte inteira (floor function)
def f4(x):
    return np.floor(x)

# Definindo os valores de x
x_cont = np.linspace(-10, 10, 400)
x_disc = np.linspace(-10, 10, 400)

# Criando figuras e eixos para os gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plotando as funções contínuas
axs[0, 0].plot(x_cont, f1(x_cont), 'r')
axs[0, 0].set_title('Função Contínua $f(x) = x^2$')

axs[0, 1].plot(x_cont, f2(x_cont), 'b')
axs[0, 1].set_title('Função Contínua $f(x) = \sin(x)$')

# Plotando as funções descontínuas
axs[1, 0].plot(x_disc, f3(x_disc), 'g')
axs[1, 0].set_title('Função Descontínua: Heaviside')

axs[1, 1].plot(x_disc, f4(x_disc), 'm')
axs[1, 1].set_title('Função Descontínua: Parte inteira $\\lfloor x \\rfloor$')

# Ajustando o layout
plt.tight_layout()

# Mostrando os gráficos
plt.show()
```

### Continuidade ajeitada

```python
import matplotlib.pyplot as plt
import numpy as np

# Função contínua 1: f(x) = x^2
def f1(x):
    return x ** 2

# Função contínua 2: f(x) = sin(x)
def f2(x):
    return np.sin(x)

# Função descontínua 1: Função de Heaviside
def f3(x):
    return np.heaviside(x, 0.5)

# Função descontínua 2: Parte inteira (floor function)
def f4(x):
    return np.floor(x)

# Definindo os valores de x
x_cont = np.linspace(-10, 10, 400)
x_disc = np.linspace(-10, 10, 400)

# Criando figuras e eixos para os gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plotando as funções contínuas
axs[0, 0].plot(x_cont, f1(x_cont), 'r')
axs[0, 0].set_title('Função Contínua $f(x) = x^2$')

axs[0, 1].plot(x_cont, f2(x_cont), 'b')
axs[0, 1].set_title('Função Contínua $f(x) = \sin(x)$')

# Plotando as funções descontínuas
# Para evitar a conexão entre os pontos descontínuos, utilizaremos o parâmetro 'drawstyle'
axs[1, 0].plot(x_disc, f3(x_disc), 'g', linestyle='none', marker='.')
axs[1, 0].set_title('Função Descontínua: Heaviside')

axs[1, 1].plot(x_disc, f4(x_disc), 'm', linestyle='none', marker='.')
axs[1, 1].set_title('Função Descontínua: Parte inteira $\\lfloor x \\rfloor$')

# Ajustando o layout
plt.tight_layout()

# Mostrando os gráficos
plt.show()
fig.savefig('graficos.png')
```
