```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')


limit_exp = lambda func, coef_ind, valor_limite: limit(func, coef_ind, valor_limite)
```

## Problema 1

$$\lim_{x \to 3} (x^2 - 9)/(x - 3)$$

## Problema 2

$$\lim_{x \to 2} \frac{x^3 - 8}{x^2 - 4}$$

## Problema 3

$$\lim_{x \to -1} \frac{x^3 + 2x^2 - x - 2}{x + 1}$$$$

```python
#1ª função
print(limit_exp((x**2 - 9)/(x-3), x, 3)) 

#2ª função
print(limit_exp((x**3-8)/(x**2-4), x, 2))

#3ª função
print(limit_exp((x**3+2*x**2-x-2)/(x+1), x, -1))
```
