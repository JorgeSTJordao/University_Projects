```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = symbols('x')
w = symbols('w')
```

```python
limit_expr = limit(((1-cos(w))/w), w, oo)
limit_expr
```

## Trabalhando com infinitos

```python
x = np.linspace(-20, 20)
y = (x-3)/(7-x) 

#Expressão
limit_expr = limit((w-3)/(7-w) , w, oo)
print(limit_expr)

plt.axvline(7, color='red', linewidth=2, linestyle='dashed')
plt.plot(x, y)
```

```python
x = np.linspace(-20, 20)
y = (x-1)/(x**2 + 5*x - 6)

#Expressão
limit_expr = limit((w-1)/(w**2 + 5*w - 6) , w, oo)
print(limit_expr)

#Gráfico
plt.plot(x, y)
```

```python
x = np.linspace(-20, 20)
y = 1/(x**2+5*x)

#Expressão
limit_expr = limit(1/(w**2+5*w) , w, 0)
print(limit_expr)

#Gráfico
plt.plot(x, y)
```
